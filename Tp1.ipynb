```python
from bs4 import BeautifulSoup
import requests
import pandas as pd

url = "https://books.toscrape.com/"
response = requests.get(url)
soup = BeautifulSoup(response.text, "html.parser")

books = []
for book in soup.find_all("article", class_="product_pod")[:10]:
    title = book.h3.a["title"]
    price = book.find("p", class_="price_color").text
    books.append({"title": title, "price": price})

df_scraping = pd.DataFrame(books)
df_scraping["source"] = "Scraping"
print("Scraping réussi:", df_scraping.shape)
df_scraping.head()
```

```python
import pandas as pd
import json
import sqlite3
from bs4 import BeautifulSoup
import requests

# 1. CSV 
df_csv = pd.DataFrame({
    "school": ["GP","GP","GP","GP","MS","MS","MS","MS"],
    "sex":    ["F","M","F","M","F","M","F","M"],
    "age":    [18, 17, 15, 16, 19, 20, 17, 18],
    "address":["U","U","U","R","U","R","U","R"],
    "G1":     [5,  6, 10, 15, 12,  8, 14, 11],
    "G2":     [6,  5, 10, 14, 13,  9, 13, 12],
    "G3":     [6,  6, 10, 15, 12, 10, 14, 11]

})
df_csv["source"] = "CSV"

# 2. JSON
with open(r"C:\Users\PRIMO TEK\anaconda3\envs\Tp!_Kdd\Data\TP1.json", "r") as f:
    json_data = json.load(f)
df_json = pd.DataFrame(json_data["students"])
df_json["source"] = "JSON"

# 3. SQL
conn = sqlite3.connect(r"C:\Users\PRIMO TEK\anaconda3\envs\Tp!_Kdd\Data\students.db")
df_sql = pd.read_sql("SELECT * FROM grades", conn)
df_sql["source"] = "SQL"

# 4. Scraping
url = "https://books.toscrape.com/"
response = requests.get(url)
soup = BeautifulSoup(response.text, "html.parser")
books = []
for book in soup.find_all("article", class_="product_pod")[:10]:
    title = book.h3.a["title"]
    price = book.find("p", class_="price_color").text
    books.append({"title": title, "price": price})
df_scraping = pd.DataFrame(books)
df_scraping["source"] = "Scraping"

# 5 Fusion
df_final = pd.concat([df_csv, df_json, df_sql, df_scraping], ignore_index=True)
print("✅ Total:", df_final.shape)
df_final.head()
```
